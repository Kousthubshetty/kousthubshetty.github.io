```python
import pandas as pd
```

```python
df = pd.read_csv('hprice.csv')
```

```python
df
```

```python
X = df['total_sqft_int']
```

```python
X
```

```python
Y = df['price']
```

```python
Y
```

```python
import numpy as np
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
n = len(X)
numer=0
denom=0
for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
m=numer/denom
c=mean_y-(m*mean_x)
```

```python
m
```

```python
c
```

```python
rmse = 0
for i in range(n):
    y_pred = c + m * X[i]
    rmse += (Y[i] - y_pred) ** 2
    
rmse = np.sqrt(rmse/n)
print("Root Mean Square Error is",rmse)
```

```python
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = c + m * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score",r2)
```
