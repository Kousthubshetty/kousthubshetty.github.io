```python
import pandas as pd
```

```python
df = pd.read_csv('hprice.csv')
```

```python
df
```

```python
X = df['total_sqft_int']
```

```python
X
```

```python
Y = df['price']
```

```python
Y
```

```python
import numpy as np
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
n = len(X)
numer=0
denom=0
for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
m=numer/denom
c=mean_y-(m*mean_x)
```

```python
m
```

```python
c
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (10.0, 5.0)


y = m * X + c

# Ploting Line
plt.plot(X, y, color='blue', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='green', label='Scatter data')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
rmse = 0
for i in range(n):
    #y_pred = c + m * X[i]
    rmse += (Y[i] - y[i]) ** 2
    
rmse = np.sqrt(rmse/n)
print("Root Mean Square Error is",rmse)
```

```python
ss_tot = 0
ss_res = 0
for i in range(n):
    #y_pred = c + m * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y[i]) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score",r2)
```
